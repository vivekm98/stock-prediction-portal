```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Type of Plots

### Line Plots

```python
x = [1,2,4,5,6]
y = [10,20,30,40,50]
```

```python
plt.figure(figsize=(15,6))
plt.plot(x,y);
plt.title("simple Line")
plt.xlabel('x Axis')
plt.ylabel('Y Axis')
plt.grid(True);
```

## Bar chart

```python
months = ['jan','feb','mar','apr']
sells = [1000,3000,2000,4000]
```

```python

plt.bar(months,sells)
plt.title('Monthly Sells')
plt.xlabel('Month')
plt.ylabel('Sells')
plt.grid(True);
```

## Histogram 

```python
marks = np.random.randint(0,100,50)
marks
```

```python
plt.hist(marks);
plt.grid(True);
```

## Scatter Plots

```python
height = [120,140,130,160]
weight = [40,30,50,60]
```

```python
plt.scatter(height,weight);
plt.title("Heigh X Weight")
plt.xlabel("height")
plt.ylabel("weight")
plt.grid(True);
```

## Subplots and Legends

```python
days = [1,2,3,4,5]
origional_price = [10,20,25,30,40]
predicted_price = [8,21,25,38,40]
```

```python
plt.plot(days,origional_price,'b-o',label="origional price")
plt.plot(days,predicted_price,'r-o',label="prdicted price")
plt.legend()
plt.title("price prediction")
plt.xlabel("days")
plt.ylabel("price")
plt.grid(True);
plt.savefig('testex.png',dpi=300);
```

## Subplot

```python
plt.subplot(3,1,1)
plt.plot(days,origional_price,'b-o',label="origional price")
```
